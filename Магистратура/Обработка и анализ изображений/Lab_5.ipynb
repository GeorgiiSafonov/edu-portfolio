```python
from copy import deepcopy
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
def print_img(img, title = '', shape = (12,9)):
  plt.figure(figsize = shape)
  plt.title(title)
  plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

def print_img_grey(img, title = '', shape = (12,9)):
  plt.figure(figsize = shape)
  plt.title(title)
  plt.imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2RGB))


def reverse_shape(shape):
  result = []
  for i in range(len(shape)):
    result.append(shape[len(shape)-i-1])
  return(tuple(result))  
```

```python
th, img_1 = cv2.threshold(cv2.imread(root_folder+"table.png",0), 150, 255, cv2.THRESH_BINARY_INV)
```

```python
print_img(img_1)
```

```python
th, img_2 = cv2.threshold(cv2.imread(root_folder+"text.jpg",0), 150, 255, cv2.THRESH_BINARY_INV)
```

```python
print_img(img_2)
```

```python
th, img_3 = cv2.threshold(cv2.imread(root_folder+"bw.jpg",0), 180, 255, cv2.THRESH_BINARY)
```

```python
print_img(img_3)
```

##Эрозия

```python
kernel_1 = np.ones((2,2), np.uint8)
erosion_1 = cv2.erode(img_1, kernel_1, iterations = 1)
print_img(np.hstack((img_1,erosion_1)))
```

```python
kernel_2 = np.ones((2,2), np.uint8)
erosion_2 = cv2.erode(img_2, kernel_2, iterations = 1)
print_img(np.hstack((img_2,erosion_2)))
```

```python
kernel_3 = np.ones((4,4), np.uint8)
erosion_3 = cv2.erode(img_3, kernel_3, iterations = 1)
print_img(np.hstack((img_3,erosion_3)))
```

## Дилатация

```python
dilation_1 = cv2.dilate(erosion_1, kernel_1,iterations = 1)
print_img(np.hstack((img_1,erosion_1,dilation_1)),'',(24,18))
```

```python
dilation_2 = cv2.dilate(erosion_2, kernel_2,iterations = 1)
print_img(np.hstack((img_2,erosion_2,dilation_2)),'',(24,18))
```

```python
dilation_3 = cv2.dilate(erosion_3, kernel_3,iterations = 1)
print_img(np.hstack((img_3, erosion_3,dilation_3)),'',(24,18))
```

## Открытие

```python
opening_1 = cv2.morphologyEx(img_1, cv2.MORPH_OPEN, kernel_1)
print_img(np.hstack((img_1,opening_1)))
```

```python
opening_2 = cv2.morphologyEx(img_2, cv2.MORPH_OPEN, kernel_2)
print_img(np.hstack((img_2,opening_2)))
```

```python
opening_3 = cv2.morphologyEx(img_3, cv2.MORPH_OPEN, kernel_3)
print_img(np.hstack((img_3,opening_3)))
```

## Закрытие

```python
closing_1 = cv2.morphologyEx(img_1, cv2.MORPH_CLOSE, kernel_1)
print_img(np.hstack((img_1,closing_1)))
```

```python
closing_2 = cv2.morphologyEx(img_2, cv2.MORPH_CLOSE, kernel_2)
print_img(np.hstack((img_2,closing_2)))
```

```python
closing_3 = cv2.morphologyEx(img_3, cv2.MORPH_CLOSE, kernel_3)
print_img(np.hstack((img_3,closing_3)))
```

## Морфологический Градиент

```python
gradient_1 = cv2.morphologyEx(img_1, cv2.MORPH_GRADIENT, kernel_1)
print_img(np.hstack((img_1,gradient_1)),'',(32,27))
```

```python
gradient_2 = cv2.morphologyEx(img_2, cv2.MORPH_GRADIENT, kernel_2)
print_img(np.hstack((img_2,gradient_2)),'',(32,27))
```

```python
gradient_3 = cv2.morphologyEx(img_3, cv2.MORPH_GRADIENT, kernel_3)
print_img(np.hstack((img_3,gradient_3)),'',(32,27))
```
