```python
from copy import deepcopy
from random import randint
import cv2
import numpy as np
import pickle
from progress.bar import IncrementalBar
import matplotlib.pyplot as plt
from numpy.random import randint
import pickle
```

```python
def print_img(img, title = '', shape = (12,9)):
  plt.figure(figsize = shape)
  plt.title(title)
  plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

rho = lambda coords_1, coords_2: pow((coords_1[0]-coords_2[0])**2 + (coords_1[1]-coords_2[1])**2, 0.5)
y = lambda x: x*0.75
x = lambda y:y/0.75

def generate_point_up(image_shape, border_width, size):
  while True:
    x_ = randint(x(size+border_width),image_shape[1]-size)
    if size < (y(x_) - border_width):
      y_ = randint(size,y(x_) - border_width)
      break
  return([x_,y_])

def generate_point_down(image_shape, border_width, size):
  x_ = randint(size, x(image_shape[0]-(size+border_width)))
  y_ = randint(y(x_)+border_width, image_shape[0]-size)
  return([x_, y_+size])
```
